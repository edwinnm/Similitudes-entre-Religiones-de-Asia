```python
import pandas as pd
```

```python
df_all_labelled  = pd.read_csv("AllBooks_baseline_DTM_Labelled.csv")
df_all_unlabelled = pd.read_csv("AllBooks_baseline_DTM_Unlabelled.csv")
file = open("Complete_data .txt", "r")
docs = []
cont = 0
for line in file:
    stripped_line = line.strip()
    if cont%2 != 0:
        docs.append(stripped_line)
    cont = cont+1

file.close()
```

```python
print("Total de Labelled:",df_all_labelled.shape)
print("Total de Unlabelled:",df_all_unlabelled.shape)
print("Total de text:",len(docs))
```
